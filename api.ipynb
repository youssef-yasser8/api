```python
import requests
import csv
import pandas as pd
```

```python
response=requests.get(r"https://coronavirus.m.pipedream.net/")
```

```python
response
```

```python
covid=response.json()
```

```python
covid.keys()
```

```python
covid['rawData']
```

```python
with open ("covid.csv","w",newline="") as f:
    writer=csv.DictWriter(f,fieldnames=['Deaths','Confirmed','Country_Region'])
    writer.writeheader()
    for co in covid['rawData'] :
        Deaths=co['Deaths']
        Confirmed=co['Confirmed']
        Country_Region=co['Country_Region']
        writer.writerow({'Deaths':Deaths ,'Confirmed':Confirmed ,'Country_Region':Country_Region})


        
        
                          
```
